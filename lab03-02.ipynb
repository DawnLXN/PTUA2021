```python
import random
import math
import numpy as np

pi = math.pi

def extended_simulation_pi(convergence_criterion, sentinel_value):
    n = 0 
    d = 0 
    simulating = 0 
    dif = 0
    s = 0
    dif_pfh = 0
    
    while simulating < sentinel_value:
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1.0:
            n += 1               # draws in the round
        d += 1                   # all draws
        ratio = 4 * n / d
        dif = abs(ratio-pi) / pi
        if dif <= convergence_criterion:
            print("Draws needed: ", d, "; estimated pi:", ratio, ";dif: ",dif)
            simulating+=1
            s += ratio
            dif_pfh += dif ** 2
        
    if simulating == sentinel_value:
        mean = s / sentinel_value
        sd = dif_pfh / sentinel_value
        print("The number of draws reaches the sentinel value.")
        print("The average number =", mean)
        print("The standard deviation = ", sd)

        

# convergence criteria= 0.01
extended_simulation_pi(0.01,10)

# convergence criteria= 0.001
extended_simulation_pi(0.001,10)

# convergence criteria= 0.0001
extended_simulation_pi(0.0001,10)

# convergence criteria= 0.00001
extended_simulation_pi(0.00001,10)
```
